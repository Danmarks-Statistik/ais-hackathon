```python
headers=['country', 'mca', 'unLocode', 'name', 'nuts3', 'lon', 'lat']
a = pd.read_excel(r'K:\125000\HAVNEANLOB\LOCODE\Eurostat\extended_unlocode-2018.xlsx',
                  sheet_name='2018 extended list of ports',
                  skiprows=1, header=None, convert_floats=False,
                  usecols=[0,1,2,3,29,33,34], 
                  names=headers)
a.dropna(inplace=True)
b = a[a['country']=='DK']
b.reset_index(drop=True, inplace=True)
```

```python
# print(b.head(2))
# type(b.lon[0])

b['lon'] = b['lon'].apply(float)
b['lat'] = b['lat'].apply(float)
# print(b.head(2))
# type(b.flon[0])

b.to_csv('statistiske havne dk.csv', index=False)
```
